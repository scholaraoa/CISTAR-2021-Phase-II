```python
import pandas as pd
import numpy as np
```

```python
df_solar = pd.read_csv('Solar Supply Data Final')
df_wind = pd.read_csv('Wind Supply Data Final')
```

```python
## Solar Generation, Realizable Capacity

df_solar['generation_mwh'] = list(map(lambda x,y:(365*24*x*y),df_solar['capacity_mw'],df_solar['capacity_factor']))
df_solar['realizable_mw'] = list(map(lambda x,y:x*y,df_solar['capacity_mw'],df_solar['capacity_factor']))
df_solar['realizable per land'] = list(map(lambda x,y:x/y,df_solar['realizable_mw'],df_solar['area_sq_km']))
df_solar.sort_values(by=['realizable per land'], inplace=True, ascending=False) # select energy dense locations
df_solar_50 = df_solar.head(50)
df_solar_20 = df_solar.head(20)
```

```python
## Wind Realizable Capacity

df_wind['realizable_mw'] = list(map(lambda x,y:x*y,df_wind['capacity_mw'],df_wind['capacity_factor']))
df_wind['realizable per land'] = list(map(lambda x,y:x/y,df_wind['realizable_mw'],df_wind['area_sq_km']))
df_wind.sort_values(by=['realizable per land'], inplace=True, ascending=False) # select energy dense locations
df_wind_50 = df_wind.head(50)
df_wind_20 = df_wind.head(20)
```
